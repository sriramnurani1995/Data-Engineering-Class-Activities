```python
from datetime import datetime, timedelta

def decode_timestamp(opd_date, act_time):
    date_time = datetime.strptime(opd_date, "%d%b%Y:%H:%M:%S")
    time_delta = timedelta(seconds=act_time)
    return date_time + time_delta

df['TIMESTAMP'] = df.apply(lambda row: decode_timestamp(row['OPD_DATE'], row['ACT_TIME']), axis=1)
df
```

```python
df_timestamp = df.drop(columns=['OPD_DATE', 'ACT_TIME'])
df_timestamp
```

```python

df_timestamp['dMETERS'] = df_timestamp['METERS'].diff()
df_timestamp['dTIMESTAMP'] = df_timestamp['TIMESTAMP'].diff().dt.total_seconds()

df_timestamp['SPEED'] = df_timestamp.apply(
    lambda row: row['dMETERS'] / row['dTIMESTAMP'] if row['dTIMESTAMP'] != 0 else 0,
    axis=1
)
df_timestamp.iloc[0, df_timestamp.columns.get_loc('SPEED')] = df_timestamp.iloc[1]['SPEED']

df_timestamp.drop(['dMETERS', 'dTIMESTAMP'], axis=1, inplace=True)


df_timestamp
```

```python

speed_stats = df_timestamp['SPEED'].describe()
print("Minimum speed:", round(speed_stats['min'], 3), "m/s")
print("Maximum speed:", round(speed_stats['max'], 3), "m/s")
print("Average speed:", round(speed_stats['mean'], 3), "m/s")
```

```python

exclude_columns = ['EVENT_NO_STOP', 'GPS_SATELLITES', 'GPS_HDOP']
all_columns = pd.read_csv("/content/drive/MyDrive/bc_trip/bc_veh4223_230215.csv", nrows=0).columns.tolist()
use_columns = [col for col in all_columns if col not in exclude_columns]
df1 = pd.read_csv("/content/drive/MyDrive/bc_trip/bc_veh4223_230215.csv", usecols=use_columns)

trips1 = df1.groupby('EVENT_NO_TRIP')
unique_trips1 = trips1.groups.keys()

max_speed = 0
max_speed_time = None
max_speed_location = None

all_speeds = []

for trip in unique_trips1:
    trip_data = trips1.get_group(trip).copy()
    trip_data['TIMESTAMP'] = trip_data.apply(lambda row: decode_timestamp(row['OPD_DATE'], row['ACT_TIME']), axis=1)
    trip_data = trip_data.drop(columns=['OPD_DATE', 'ACT_TIME'])

    trip_data['dMETERS'] = trip_data['METERS'].diff()
    trip_data['dTIMESTAMP'] = trip_data['TIMESTAMP'].diff().dt.total_seconds()
    trip_data['SPEED'] = trip_data.apply(lambda row: row['dMETERS'] / row['dTIMESTAMP'] if row['dTIMESTAMP'] != 0 else 0, axis=1)
    if len(trip_data) > 1:
        trip_data.iloc[0, trip_data.columns.get_loc('SPEED')] = trip_data.iloc[1]['SPEED']
    else:
        trip_data.iloc[0, trip_data.columns.get_loc('SPEED')] = 0

    trip_data.drop(['dMETERS', 'dTIMESTAMP'], axis=1, inplace=True)
    all_speeds.extend(trip_data['SPEED'].tolist())
    current_max_speed = trip_data['SPEED'].max()
    if current_max_speed > max_speed:
        max_speed = current_max_speed
        max_speed_row = trip_data[trip_data['SPEED'] == max_speed].iloc[0]
        max_speed_time = max_speed_row['TIMESTAMP']
        max_speed_location = (max_speed_row['GPS_LATITUDE'], max_speed_row['GPS_LONGITUDE'])

median_speed = pd.Series(all_speeds).median()

print("Maximum Speed:", max_speed, "m/s")
print("Occurred at:", max_speed_time, "at location", max_speed_location)
print("Median Speed:", median_speed, "m/s")
```

```python
import pandas as pd


exclude_columns = ['EVENT_NO_STOP', 'GPS_SATELLITES', 'GPS_HDOP']
all_columns = pd.read_csv("/content/drive/MyDrive/bc_trip/bc_230215.csv", nrows=0).columns.tolist()
use_columns = [col for col in all_columns if col not in exclude_columns]
df2 = pd.read_csv("/content/drive/MyDrive/bc_trip/bc_230215.csv", usecols=use_columns)


trips2 = df2.groupby('EVENT_NO_TRIP')

max_speed = 0
max_speed_details = None

mean_speeds = {}

for trip in trips2.groups.keys():
    trip_data = trips2.get_group(trip).copy()
    trip_data['TIMESTAMP'] = trip_data.apply(lambda row: decode_timestamp(row['OPD_DATE'], row['ACT_TIME']), axis=1)
    trip_data = trip_data.drop(columns=['OPD_DATE', 'ACT_TIME'])

    trip_data['dMETERS'] = trip_data['METERS'].diff()
    trip_data['dTIMESTAMP'] = trip_data['TIMESTAMP'].diff().dt.total_seconds()
    trip_data['SPEED'] = trip_data.apply(lambda row: row['dMETERS'] / row['dTIMESTAMP'] if row['dTIMESTAMP'] != 0 else 0, axis=1)
    if len(trip_data) > 1:
        trip_data.iloc[0, trip_data.columns.get_loc('SPEED')] = trip_data.iloc[1]['SPEED']
    else:
        trip_data.iloc[0, trip_data.columns.get_loc('SPEED')] = 0
    trip_data.drop(['dMETERS', 'dTIMESTAMP'], axis=1, inplace=True)


    mean_speed = trip_data['SPEED'].mean()
    mean_speeds[(trip_data.iloc[0]['VEHICLE_ID'], trip)] = mean_speed


    current_max_speed = trip_data['SPEED'].max()
    if current_max_speed > max_speed:
        max_speed = current_max_speed
        max_speed_row = trip_data[trip_data['SPEED'] == max_speed].iloc[0]
        max_speed_details = {
            'time': max_speed_row['TIMESTAMP'],
            'location': (max_speed_row.get('GPS_LATITUDE', 'Not available'), max_speed_row.get('GPS_LONGITUDE', 'Not available')),
            'vehicle_id': max_speed_row['VEHICLE_ID']
        }


fastest_trip = max(mean_speeds, key=mean_speeds.get)
fastest_vehicle, fastest_trip_id = fastest_trip
fastest_mean_speed = mean_speeds[fastest_trip]

print("Maximum Speed:", max_speed, "m/s")
print("Maximum Speed Details:", max_speed_details)
print(f"The fastest average speed was {fastest_mean_speed:.2f} m/s by vehicle {fastest_vehicle} on trip {fastest_trip_id}.")
```
